```python
# Import required libraries
from yahoo_fin import news
from finvizfinance.quote import finvizfinance
from pytickersymbols import PyTickerSymbols
import logging
from datetime import datetime, timedelta, timezone
import time
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import matplotlib.pyplot as plt
import pandas as pd
import requests
import csv
import pytz
import os

# Download the VADER lexicon and punkt tokenizer
nltk.download('vader_lexicon')
nltk.download('punkt')

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger()

# Function to get S&P 500 tickers and their company names using PyTickerSymbols
def get_sp500_tickers_and_names():
    print("Fetching S&P 500 tickers and company names...")
    stock_data = PyTickerSymbols()
    sp500_data = stock_data.get_stocks_by_index('S&P 500')
    tickers_to_names = {entry['symbol']: entry['name'] for entry in sp500_data}
    print(f"Retrieved {len(tickers_to_names)} tickers and company names.")
    return tickers_to_names

def fetch_stock_data(tickers):
    try:
        # Construct the API URL based on the provided tickers
        tickers_string = ','.join(tickers)
        api_url = f"https://elite.finviz.com/export.ashx?v=111&t={tickers_string}&auth=3f2b0514-0774-4c2d-b2af-a9a69d2c7fc5"

        # Fetch the data
        response = requests.get(api_url)
        if response.status_code == 200:
            return response.content
        else:
            print(f"Failed to fetch data. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error fetching data: {str(e)}")
        return None

def append_csv(file_path, new_data):
    new_rows = list(csv.reader(new_data.decode('utf-8').splitlines()))
    header = new_rows[0]

    # Ensure that the 'Price' column is present in the data
    if "Price" in header:
        price_index = header.index("Price")
        ticker_index = header.index("Ticker")
    else:
        print("No 'Price' column found in the data. Exiting.")
        return

    # Add timestamp column
    header = ["Timestamp", "Ticker", "Price"]

    # Convert current time to EDT
    utc_now = datetime.now(timezone.utc)
    eastern = pytz.timezone('US/Eastern')
    current_time_edt = utc_now.astimezone(eastern).strftime("%Y-%m-%d %H:%M:%S")

    rows_to_append = []
    for row in new_rows[1:]:
        ticker = row[ticker_index]
        price = row[price_index]
        rows_to_append.append([current_time_edt, ticker, price])

    file_exists = os.path.isfile(file_path)

    with open(file_path, 'a', newline='') as file:
        writer = csv.writer(file)
        if not file_exists:
            writer.writerow(header)
        writer.writerows(rows_to_append)
        print("Price data appended to CSV file successfully.")

def get_latest_news(ticker: str, last_checked_time: datetime, source: str = 'yf') -> list:
    news_list = []
    try:
        if source == 'yf':
            logger.info(f'Fetching latest news for {ticker} from Yahoo! Finance')
            news_data = news.get_yf_rss(ticker)
            for article in news_data:
                article_datetime = datetime.strptime(article["published"], "%a, %d %b %Y %H:%M:%S %z")
                if article_datetime > last_checked_time:
                    news_list.append({
                        "ticker": ticker,
                        "title": article["title"],
                        "summary": article.get("summary", None),
                        "url": article["link"],
                        "time": article["published"]
                    })
        elif source == 'finviz':
            logger.info(f'Fetching latest news for {ticker} from Finviz')
            news_data = finvizfinance(ticker).ticker_news()
            for index, row in news_data.iterrows():
                # Assuming row["Date"] is already a Timestamp object
                article_datetime = row["Date"]

                # If it's not a timezone-aware timestamp, convert it
                if article_datetime.tzinfo is None:
                    article_datetime = article_datetime.tz_localize('UTC')

                if article_datetime > last_checked_time:
                    news_list.append({
                        "ticker": ticker,
                        "title": row["Title"],
                        "summary": None,
                        "url": row["Link"],
                        "time": row["Date"]
                    })
        else:
            raise ValueError("Invalid news source. Please choose either 'yf' or 'finviz'.")

        return news_list

    except Exception as e:
        logger.error(f"An error occurred: {e}")
        return []

# Function to summarize news content
def summarize_content(text, sentence_count=2):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = LsaSummarizer()
    summary = summarizer(parser.document, sentence_count)
    return ' '.join(str(sentence) for sentence in summary)

# Function to perform sentiment analysis using VADER
def perform_sentiment_analysis(text):
    sid = SentimentIntensityAnalyzer()
    sentiment_scores = sid.polarity_scores(text)
    return sentiment_scores

def fetch_price_from_csv(file_name, ticker):
    try:
        with open(file_name, 'r') as file:
            reader = csv.DictReader(file)
            prices = []
            for row in reader:
                if row['Ticker'] == ticker:
                    prices.append(float(row['Price']))
            if prices:
                return prices
            else:
                print(f"Ticker {ticker} not found in the CSV file.")
                return None
    except Exception as e:
        print(f"Error reading CSV file: {e}")
        return None

# Function to clear the CSV file
def clear_csv(file_path):
    header = ["Timestamp", "Ticker", "Price"]
    with open(file_path, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)
    print("CSV file cleared successfully.")


# Main loop to fetch news, perform sentiment analysis, monitor price, and plot results
results = []

# Get S&P 500 tickers and their company names
sp500_tickers = get_sp500_tickers_and_names()

# Monitoring period in minutes
monitoring_period = 5

file_path = "export_historical5.csv"

# Initialize the plot outside the loop to maintain existing points
fig, ax = plt.subplots(figsize=(10, 8))

eastern = pytz.timezone('US/Eastern')

now = datetime.now(eastern)
start_time = now.replace(hour=5, minute=0, second=0, microsecond=0)
end_time = now.replace(hour=18, minute=30, second=0, microsecond=0)

while True:
    current_time = datetime.now(timezone.utc)
    last_checked_time = current_time - timedelta(minutes=5)  # Look for news in the last 5 minutes

    # Clear the CSV file at the start of each cycle
    clear_csv(file_path)

    # Dictionary to store tickers with recent news
    recent_news_tickers = {}

    # Fetch and process news for each ticker
    for ticker in sp500_tickers.keys():
        news_articles = get_latest_news(ticker, last_checked_time, source='yf')
        if news_articles:
            recent_news_tickers[ticker] = news_articles

    if not recent_news_tickers:
        print("No news found for any ticker in the last 5 minutes.")
    else:
        print(f"Found news for {len(recent_news_tickers)} tickers: {list(recent_news_tickers.keys())}")
        # Monitor stock prices over the next 5 minutes
        initial_prices = {}
        final_prices = {}

        for i in range(5):
            print(f"Fetching price data iteration {i + 1}...")
            if datetime.now(eastern) >= start_time and datetime.now(eastern) < end_time:
                tickers_to_fetch = list(recent_news_tickers.keys())  # Get the list of tickers with news
                data = fetch_stock_data(tickers_to_fetch)
                if data:
                    append_csv(file_path, data)
                else:
                    print("Failed to fetch data. Skipping this update.")

            # Wait for 1 minute before the next fetch
            time.sleep(60)

        print("Calculating price changes...")
        # Fetch initial prices
        for ticker in recent_news_tickers.keys():
            prices = fetch_price_from_csv(file_path, ticker)
            if prices:
                initial_prices[ticker] = prices[0]  # Take the first price as the initial price

        # Calculate final prices and price changes
        for ticker in recent_news_tickers.keys():
            prices = fetch_price_from_csv(file_path, ticker)
            if prices:
                final_price = prices[-1]  # Take the last price as the final price
                initial_price = initial_prices.get(ticker)
                if initial_price is not None:
                    price_change = final_price - initial_price
                    print(f"Price change for {ticker} over 5 minutes: {price_change}")

                    print(f"Plotting data for {ticker}...")
                    # Analyze news and plot sentiment vs. price change
                    news_list = recent_news_tickers[ticker]
                    for news_item in news_list:
                        print(f"News for {ticker} ({sp500_tickers[ticker]}): {news_item['title']}")

                        if news_item['summary']:
                            summary = summarize_content(news_item['summary'])
                            sentiment = perform_sentiment_analysis(summary)

                            print(f"Sentiment Analysis for {ticker}: {sentiment}")

                            # Append to results for persistent data
                            results.append({
                                'Ticker': ticker,
                                'Sentiment': sentiment['compound'],
                                'Price_Change': price_change
                            })

        # Plotting data
        plt.figure(figsize=(10, 8))
        ax = plt.gca()
        ax.set_xlabel('Sentiment Score')
        ax.set_ylabel('Price Change')
        ax.set_title('Stock Sentiment vs Price Change')
        ax.set_xlim([-1, 1])
        initial_ylim = [-1, 1]
        ax.set_ylim(initial_ylim)
        ax.axhline(y=0, color='red', linestyle='--')  # Horizontal line at y=0
        ax.axvline(x=0, color='red', linestyle='--')  # Vertical line at x=0
        ax.grid(True)

        # Re-plot all data points in `results`
        for result in results:
            ax.scatter(result['Sentiment'], result['Price_Change'], color='blue')
            ax.text(result['Sentiment'], result['Price_Change'], result['Ticker'], fontsize=8, ha='center', va='bottom')

        plt.show()

    print("Completed a 5-minute cycle. Sleeping before starting the next cycle.")
    time.sleep(60)  # Sleep for 1 minute before starting the next 5-minute cycle
```
